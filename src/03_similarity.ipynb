#### We create a dataset to store the accuracies

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv(r'..\data\interim\clean_movies.csv')
ratings = pd.read_csv(r'..\data\raw\ratings.csv')
```

```python
movies.head()
```

```python
ratings.head()
```

```python
data = ratings.pivot_table(index='id', columns='movies_id', values='Ratings', aggfunc='first')

# for i in data.index:
#     print(i)
```

```python
data.iloc[:, :5].head(5)
```

```python
data=data.fillna(0)
data.iloc[:, :5].head(5)
```

```python
data
```

```python
data.columns = movies['Film'].tolist()
```

```python
data.head()
```

```python
user_similarity = cosine_similarity(data)
```

```python
data.to_csv(r'..\data\processed\pivot_mtx.csv',index=False)
```
